# Implement Queue using Stacks [Easy]

https://leetcode.com/problems/implement-queue-using-stacks/description/

Implement a first in first out (FIFO) queue using only two stacks. The implemented queue should support all the functions of a normal queue (**push**, **peek**, **pop**, and **empty**).

Implement the **MyQueue** class:

- **void push(int x)** Pushes element x to the back of the queue.
- **int pop()** Removes the element from the front of the queue and returns it.
- **int peek()** Returns the element at the front of the queue.
- **boolean empty()** Returns true if the queue is empty, false otherwise.

Notes:

- You must use only standard operations of a stack, which means only push to top, peek/pop from top, size, and is empty operations are valid.
- Depending on your language, the stack may not be supported natively. You may simulate a stack using a list or deque (double-ended queue) as long as you use only a stack's standard operations.
 

## Example 1:
```
Input
["MyQueue", "push", "push", "peek", "pop", "empty"]
[[], [1], [2], [], [], []]
Output
[null, null, null, 1, 1, false]

Explanation
MyQueue myQueue = new MyQueue();
myQueue.push(1); // queue is: [1]
myQueue.push(2); // queue is: [1, 2] (leftmost is front of the queue)
myQueue.peek(); // return 1
myQueue.pop(); // return 1, queue is [2]
myQueue.empty(); // return false
 ```

## Constraints:

- 1 <= x <= 9
- At most 100 calls will be made to push, pop, peek, and empty.
- All the calls to pop and peek are valid.
 

**Follow-up:** Can you implement the queue such that each operation is amortized O(1) time complexity? In other words, performing n operations will take overall O(n) time even if one of those operations may take longer.

````python
class StackArray:
    def __init__(self, capacity):
        self.capacity = capacity
        self.items = [None] * capacity
        self.size = 0


    def push(self, item:any) -> None:
        if self.size == self.capacity:
            raise Exception('Stack is full!!')
        
        self.items[self.size] = item
        self.size += 1
    
    def pop(self) -> any:
        if self.size == 0:
            return None
        
        self.size -= 1
        item = self.items[self.size]
        self.items[self.size] = None
        
        return item
    
    def peek(self) -> any:
        if self.size == 0:
            return None
        
        return self.items[self.size - 1]
        
    def __len__(self):
        return self.size
    
    def __repr__(self):
        l = []
        for i in range(self.size):
            idx = (self.size - 1) - i
            l.append(self.items[idx])

        return str(l)

class MyQueue:
    def __init__(self, capacity=100):
        self.capacity = capacity
        self.forwards = StackArray(capacity)
        self.backwards = StackArray(capacity)

    def push(self, x: int) -> None:
        """Need to push to the back"""
        for i in range(len(self.backwards)):
            self.forwards.push(self.backwards.pop())

        self.forwards.push(x)
        
    def pop(self) -> int:
        """Pop from the front"""
        for i in range(len(self.forwards)):
            self.backwards.push(self.forwards.pop())

        return self.backwards.pop()

    def peek(self) -> int:
        """Peek first element"""
        for i in range(len(self.forwards)):
            self.backwards.push(self.forwards.pop())

        return self.backwards.peek()

    def empty(self) -> bool:
        if len(self.forwards) + len(self.backwards) == 0:
            return True
        
        return False
    
    def __repr__(self):
        return f"Forwards: {self.forwards}\\Backwards: {self.backwards}"


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
````

````python
class StackArray2:
    def __init__(self, capacity):
        self.capacity = capacity
        self.size = 0
        self.items = [None] * capacity

    def push(self, item:any) -> None:
        if self.size == self.capacity:
            raise Exception("The Stack is Full!!")
        
        self.items[self.size] = item
        self.size += 1

    def pop(self) -> any:
        if self.size == 0:
            return None
        
        self.size -=1
        item = self.items[self.size] 
        self.items[self.size] = None

        return item
    
    def peek(self) -> any:
        if self.size == 0:
            return None
        
        return self.items[self.size - 1]
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self):
        l = []
        for i in range(self.size):
            idx = (self.size - 1) - i
            l.append(self.items[idx])

        return str(l)
    
class MyQueue2:
    def __init__(self, capacity=100):
        self.capacity = capacity
        self.size = 0
        self.front = StackArray2(capacity)
        self.back = StackArray2(capacity)


    def push(self, item: any) -> None:
        """Queue into the back"""
        for i in range(len(self.front)):
            self.back.push(self.front.pop())

        self.back.push(item)


    def pop(self) -> any:
        """Take from the front"""
        for i in range(len(self.back)):
            self.front.push(self.back.pop())

        return self.front.pop()
    
    def peek(self) -> any:
        """Peek at the front"""

        for i in range(len(self.back)):
            self.front.push(self.back.pop())

        return self.front.peek()

    def empty(self) -> bool:
        if len(self.front) + len(self.back) == 0:
            return True
        
        return False

    def __repr__(self):
        return f"Front: {self.front}\\Back: {self.back}"
    
    
````

````python
s = StackArray2(5)
s.push(1)
s.push(2)
s.push(3)
assert len(s) == 3
assert s.peek() == 3
assert s.pop() == 3
assert len(s) == 2
assert s.pop() == 2
assert len(s) == 1
s.push(5)
assert len(s) == 2
assert s.peek() == 5


q = MyQueue2()
q.push(1)
q.push(2)
assert len(q.front) == 0
assert len(q.back) == 2
print(q)
print(q.peek())
print(q.peek())
assert q.peek() == 1
assert q.empty() == False
````
