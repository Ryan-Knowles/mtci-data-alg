# Longest Palindrome Substring [medium]

Given a string s, return the longest palindromic substring in s.

## Example 1:

Input: s = "babad"

Output: "bab"

Explanation: "aba" is also a valid answer.

## Example 2:

Input: s = "cbbd"

Output: "bb"

## Constraints:

- 1 <= s.length <= 1000
- s consist of only digits and English letters.



```python
from math import floor, ceil

class Solution:
    def isPalindrome(self, s:str, i, j) -> bool:
        """i is the starting index and j is the ending index, inclusive"""
        if i == j:
            return True

        size = (j - i) // 2

        return  s[i:i+size+1] == s[j-size:j+1][::-1]

    def longestPalindrome(self, s: str) -> str:

        # Keep track here
        doubles_len = 2
        singles_len = 1
        max_len = 1
        max_pos = 0
        is_max_doubles = False


        for i in range(len(s)):
            # Starting from i, see if we can find palindromes of increasing size
            for size in range(1, len(s)//2 + 1):
                # Check bounds of string
                if i-size < 0 or i+size >= len(s):
                    break

                # Check if palindrome
                if not self.isPalindrome(s, i-size, i+size):
                    break

                # It is a palindrome so update our metrics and continue
                singles_len = size * 2 + 1
                if singles_len > max_len:
                    max_len = singles_len
                    max_pos = i
                    is_max_doubles = False

            # Finished with checking singles, so reset it
            singles_len = 1
            
            # IF i and i+1 are the same, start finding palindromes with that as a start

            # Check bounds and equality to see if we have a doubles
            if i+1 >= len(s) or s[i] != s[i + 1]:
                continue
            
            # Same as above, just checking for doubles now
            for size in range(len(s)//2):
                # Check bounds of string
                if i-size < 0 or i+1+size >= len(s):
                    break

                # Check if palindrome
                if not self.isPalindrome(s, i-size, i+1+size):
                    break

                # It is a palindrome so update our metrics and continue
                doubles_len = size * 2 + 2
                if doubles_len > max_len:
                    max_len = doubles_len
                    max_pos = i
                    is_max_doubles = True

            # Finished with doubles, so reset len
            doubles_len = 2

        start = max_pos - (max_len - 2*int(is_max_doubles))//2
        end = max_pos + (max_len - 2*int(is_max_doubles))//2 + int(is_max_doubles)
        return s[start:end+1]





print(f"{Solution().isPalindrome('abba', 0,3) =}")
print(f"{Solution().isPalindrome('aba', 0,2) =}")
print(f"{Solution().isPalindrome('abbba', 0,4) =}")
print(f"{Solution().isPalindrome('abababa', 0,6) =}")
print(f"{Solution().isPalindrome('1abababa2', 1,7) =}")



print(f"{Solution().isPalindrome('sabad', 1,3) =}")

def test1():
    arr = [0, 1, 2, 3, 4, 4, 6, 7, 8]
    max_len = 6
    max_pos = 4
    is_doubles = True

    start = max_pos - (max_len - 2*int(is_doubles))//2
    end = max_pos + (max_len - 2*int(is_doubles))//2 + int(is_doubles)
    print(arr[start:end+1])

def test2():
    arr = [0, 1, 2, 3, 4, 3, 2, 1, 8]
    max_len = 7
    max_pos = 4
    is_doubles = False

    start = max_pos - (max_len - 2*int(is_doubles))//2
    end = max_pos + (max_len - 2*int(is_doubles))//2 + int(is_doubles)
    print(arr[start:end+1])


def test3(s, i, j):
    size = (j - i) / 2
    midpoint_l = floor(size) + i
    midpoint_r = j - floor(size)
    print(f"{midpoint_l =} {midpoint_r =}")
    print(f"{s[i:midpoint_l+1] =} == {s[midpoint_r:j+1][::-1] =}")


test3("sabad", 1, 3)
test3("sabbad", 1, 4)
test3('cbbd', 1, 2)
solution = Solution()
print(solution.longestPalindrome("aaaaa"))
```

```python
def test_solution():
    solution = Solution()
    print(solution.longestPalindrome("babad"))
    
    # Test case 1: Simple palindrome
    assert solution.longestPalindrome("babad") in ["bab", "aba"]
    
    # Test case 2: Even length palindrome
    assert solution.longestPalindrome("cbbd") == "bb"
    
    # Test case 3: Single character
    assert solution.longestPalindrome("a") == "a"
    
    # Test case 4: Two identical characters
    assert solution.longestPalindrome("aa") == "aa"
    
    # Test case 5: All characters are the same
    assert solution.longestPalindrome("aaaaa") == "aaaaa"
    
    # Test case 6: No palindrome longer than 1
    assert len(solution.longestPalindrome("abcd")) == 1
    
    # Test case 7: Long palindrome in the middle
    assert solution.longestPalindrome("xabcbay") == "abcba"
    
    # Test case 8: Palindrome at the beginning
    assert solution.longestPalindrome("abcdcba123") == "abcdcba"
    
    # Test case 9: Palindrome at the end
    assert solution.longestPalindrome("123abcdcba") == "abcdcba"
    
    # Test case 10: Multiple palindromes, return the longest
    assert solution.longestPalindrome("aabbccbbaa123454321") in ["aabbccbbaa", "123454321"]
    
    print("All test cases passed!")

# Run the tests
test_solution()
```
